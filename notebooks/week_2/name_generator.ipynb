# Generating names for search engine auditing

```python
import pandas as pd
import numpy as np
```

```python
sweeney_names = pd.read_csv("../../data/sweeney_names.csv")
sweeney_names
```

```python
last_names = ["Green", "Harris", "Hall", "Williams", "Johnson"]
```

```python
names = []
locations = [" New York", " California", " Texas", " Florida"]
initials = [" A. ", " C. ", " E. ", " M. "]

for location in locations:
    for name in sweeney_names['White Male']:
        for initial in initials:
            for lname in last_names:
                if name is not np.nan:
                    names.append({'query':name + initial + lname + location,
                                  'group':'names',
                                  'subgroup':'White Masculine'})

    for name in sweeney_names['White Female']:
        for lname in last_names:
            for initial in initials:
                if name is not np.nan:
                    names.append({'query':name + initial + lname + location,
                                  'group':'names',
                                  'subgroup':'White Feminine'})

    for name in sweeney_names['Black Female']:
        for lname in last_names:
            for initial in initials:
                if name is not np.nan:
                    names.append({'query':name + initial + lname + location,
                                  'group':'names',
                                  'subgroup':'Black Feminine'})

    for name in sweeney_names['Black Male']:
        for lname in last_names:
            for initial in initials:
                if name is not np.nan:
                    names.append({'query':name + initial + lname + location,
                                  'group':'names',
                                  'subgroup':'Black Masculine'})

```

```python
names_df = pd.DataFrame(names)
names_df

```
